<h5>removing outliers using percentile</h5>

```python
lb = df['price'].quantile(0.01) # upper bound
lb
```

```python
ub = df['price'].quantile(0.999) # lower bound 
ub
```

```python

price_range = df[(df['price'] >= lb) & (df['price'] <= ub)]
price_range
```

```python
price_range.sort_values(by='price').head(20)
```

```python
price_range.sort_values(by='price').tail(20)
```
